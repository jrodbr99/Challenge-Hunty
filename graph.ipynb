```python
import matplotlib.pyplot as plt

# Data
number_of_questions = [0, 1, 2, 3, 4, 5, 6]
number_of_candidates = [1, 4, 2, 3, 1, 2, 16]

# Create bar chart
plt.bar(number_of_questions, number_of_candidates, color='skyblue')
plt.xlabel('Number of Questions Answered')
plt.ylabel('Number of Candidates')
plt.title('Distribution of Candidates by Number of Questions Answered')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
